#### Example:1 Driver to scrape Men's kurta and kurtasets category

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import time

#Initializing chrome driver
driver = webdriver.Chrome(ChromeDriverManager().install())

page_count = 1


# Create empty dataframe to store details
df = pd.DataFrame(columns=['brand', 'product', 'price', 'rating', 'rating_counts', 'link', 'material_info'])

for i in range (1,21): 
    # Navigate to website
    url = 'https://www.myntra.com/men-ethnic-wear?f=Categories%3AKurta%20Sets%2CKurtas&sort=popularity&p=' + str(page_count)
    driver.get(url)
    page_count += 1
    # Find total number of items for the category
    total_items = int(driver.find_elements_by_class_name('title-count')[0].text.split()[1])

    # Delay the execution of the next line of code by 3 seconds
    time.sleep(3)

    # Store the ID of the original window
    original_window = driver.current_window_handle

    # Delay the execution of the next line of code by 3 seconds
    time.sleep(3)

    # Find all items on the page
    items = driver.find_elements_by_class_name('product-base')

    # Iterate through each item
    for item in items:
        # Find brand
        brand = item.find_elements_by_class_name('product-brand')[0].text

        # Find product name
        product = item.find_elements_by_class_name('product-product')[0].text

        try:
            # Find product rating
            rating = item.find_elements_by_class_name('product-ratingsContainer')[0].text
            # Find product rating counts
            rating_counts = item.find_elements_by_class_name('product-ratingsCount')[0].text
        except:
            rating = None
            rating_counts = None

        # Check if discounted price is available
        try:
            price = item.find_elements_by_class_name('product-discountedPrice')[0].text.split()[1]
        except:
            price = item.find_elements_by_class_name('product-price')[0].text.split()[1]


        # Find product link
        link = item.find_element_by_tag_name('a').get_property("href")       

        # To find material_info, click product href link 
        item.find_elements_by_tag_name('a')[0].click()

        # Switch to new tab
        # print(len(driver.window_handles))

        # Loop through until we find a new window handle
        for window_handle in driver.window_handles:
            if window_handle != original_window:
                driver.switch_to.window(window_handle)
                time.sleep(5)
                try:
                    material_info = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div[1]/main/div[2]/div[2]/div[3]/div/div[3]/p').text
                except:
                    try:
                        material_info = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div[1]/main/div[2]/div[2]/div[3]/div/div[2]/p').text
                    except:
                        material_info = None
                driver.close()
                break
        #Switch back to the old tab or window
        driver.switch_to.window(original_window)

        # Append details to dataframe
        df = df.append({'brand': brand, 'product': product , 'rating': rating, 'rating_counts': rating_counts, 'price': price, 'link': link, 'material_info': material_info}, ignore_index=True)
        
        
# Close web driver
driver.quit()


# Print dataframe
print(df)

# Create csv
df.to_csv('men_kurta_kurtasets.csv')
```
